```python
import pandas as pd
import random
from multiprocessing import Process, Value, Array, Pool
import pickle
import numpy as np
data = pd.read_excel('data/mainData.xlsx',sheet_name='Final Data workings')
feature_cols = np.array(['n2', 'o2', 'h2', 'ch4'])
thresh_dict = {'n2': 5, 'o2': 0.5, 'h2': 50, 'ch4': 500}
fitness_dict = {}
with open('fitness_dict.pkl', 'rb') as fp:
    fitness_dict = pickle.load(fp)
```

```python
def create_individuals():
    individuals = []
    for i in range(len(feature_cols)):
    
        individuals.append([feature_cols[i]])
        for j in range(i+1,len(feature_cols)):
            individuals.append(['(',feature_cols[i],'|',feature_cols[j],')'])
            individuals.append(['(',feature_cols[i],'&',feature_cols[j],')'])
    size = len(individuals)
    for i in range(size):
        for j in range(i+1,size):
            individuals.append(individuals[i]+['|']+individuals[j])
            individuals.append(individuals[i]+['&']+individuals[j])
    return individuals


def cross_over(feature1,feature2):
    feat1_op_indexs = []
    feat2_op_indexs = []

    for i in range(len(feature1)):
        if feature1[i] =='|' or feature1[i]=='&':
            feat1_op_indexs.append(i)


    for i in range(len(feature2)):
        if feature2[i] =='|' or feature2[i]=='&':
            feat2_op_indexs.append(i)
    idx1 = -1
    idx2 = -1
    if feat1_op_indexs:
        idx1 = random.choice(feat1_op_indexs)
    if feat2_op_indexs:
        idx2 = random.choice(feat2_op_indexs)
    open = 0
    close = 0
    idx1_end = -1
    idx2_end = -1
    idx1_start = 0
    idx2_start = 0
    for i in range(idx1+1,len(feature1)):
        if feature1[i]=='(':
            open+=1
        if feature1[i]==')':
            close+=1
        if close>open:
            idx1_end =i
            break
    if idx1_end==-1:
        idx1_end = len(feature1)-1
    open =0
    close = 0
    for i in range(idx1-1,len(feature1),-1):
        if feature1[i]=='(':
            open+=1
        if feature1[i]==')':
            close+=1
        if close<open:
            idx1_start =i
            break

    open =0
    close = 0
    for i in range(idx2-1,len(feature2),-1):
        if feature2[i]=='(':
            open+=1
        if feature2[i]==')':
            close+=1
        if close<open:
            idx2_start =i
            break

    open = 0
    close = 0
    for i in range(idx2+1,len(feature2)):
        if feature2[i]=='(':
            open+=1
        if feature2[i]==')':
            close+=1
        if close>open:
            idx2_end =i
            break
    if idx2_end==-1:
        idx2_end = len(feature2)-1
  
    new_feature1 = feature1[:idx1_start] + feature2[idx2_start:idx2_end+1] + feature1[idx1_end+1:]
    new_feature2 = feature2[:idx2_start] + feature1[idx1_start:idx1_end+1] + feature2[idx2_end+1:]
 
    return [new_feature1,new_feature2]
```

```python
def fitness_processor(index,feature1,output):
    stride = int(len(data)/8)
    start = index*stride
    for row in range(start,min((index+1)*stride,len(data))):
        if expression_calculator(feature1,row) =='+':
          if data.loc[row].at['Status']=='Healthy':
            output[3]+=1
          else:
            output[0]+=1
        else:
          if data.loc[row].at['Status']=='Healthy':
            output[1]+=1
          else:
            output[2]+=1
    return
def fitness_function(feature1):
    tp1 = 0
    tn1 = 0
    fp1 = 0
    fn1 = 0
    output = Array('i',[0]*4)
    processes = [Process(target=fitness_processor, args=(i,feature1,output)) for i in range(8)]
    for p in processes:
      p.start()
    for p in processes:
      p.join()
    return (output[0]/(output[0]+output[3]))*(output[1]/(output[1]+output[2]))
   


def apply_operation(col1,col2,op,row):
  if col1!='+' and col1!='-':
    if data.loc[row].at[col1] >= thresh_dict[col1]:
      col1 ='+'
    else:
      col1='-'

  if col2!='+' and col2!='-':
    if data.loc[row].at[col2] >= thresh_dict[col2]:
      col2 ='+'
    else:
      col2='-'

  if op =='|':
    if col1=='+' or col2 =='+':
      return '+'
    else:
      return '-'
  else:
    if col1=='+' and col2=='+':
      return '+'
    else:
      return '-'
  return



def expression_calculator(expression,row):
  values = []
  ops = []
  i = 0

  while i<len(expression):
    if expression[i]=='(':
      ops.append('(')
    elif expression[i]!='(' and expression[i]!=')' and expression[i]!='|' and expression[i]!='&':
      col_name = ''
      while i<len(expression) and expression[i]!='(' and expression[i]!=')' and expression[i]!='|' and expression[i]!='&':
        col_name+=expression[i]
        i+=1
      values.append(col_name)
      i-=1
    elif expression[i] ==')':
      while ops and ops[-1]!='(':
        col2 = values[-1]
        values.pop()
        col1 = values[-1]
        values.pop()
        oper = ops[-1]
        ops.pop()
        values.append(apply_operation(col1,col2,oper,row))
      if ops:
        ops.pop()
    elif expression[i] =='|' or expression[i] =='&':
      ops.append(expression[i])
    i+=1
  while ops:
        col2 = values[-1];
        values.pop()
                 
        col1 = values[-1];
        values.pop()
                 
        oper = ops[-1];
        ops.pop()
                 
        values.append(apply_operation(col1,col2,oper,row))
  if values[-1]!='+' and values[-1]!='-':
    if data.loc[row].at[values[-1]]>=thresh_dict[values[-1]]:
      values[-1] = '+'
    else:
      values[-1] = '-'
  return values[-1]
```

```python
def mutation(feature1):
  op_indexs = []
  feature = []
  for i in range(len(feature1)):
    if feature1[i] =='|' or feature1[i]=='&':
      op_indexs.append(i)
  if op_indexs:
    idx1 = random.choice(op_indexs)
    if feature1[idx1] =='|':
      feature = list(feature1)
      feature[idx1] = '&'
      
    else:
      feature = list(feature1)
      feature[idx1] = '|'
      
  return "".join(feature)

def fitness_cache(individuals):
  sum = 0
  for i in range(len(individuals)):
    if individuals[i] not in fitness_dict.keys():
      fitness_dict[individuals[i]] = fitness_function(individuals[i])
    sum+=fitness_dict[individuals[i]]
  return sum


def genetic_programming():
  gen = create_individuals()
  fitness_sum = fitness_cache(gen)
  sorted(gen,key=lambda ind: fitness_dict[ind],reverse=True)
  gen = gen[:101]
  iterations = 0
  while iterations<5:
    new_gen = gen.copy()
    for i in range(len(gen)):
      if random.randrange(9)<2:
        new_feature = mutation(gen[i])
        new_gen.append(new_feature)
      for ind2 in range(i+1,len(gen)):
          if random.randrange(9) < 3:
            new_feature1,new_feature2 = cross_over(gen[i],gen[ind2])
            new_gen.append(new_feature1)
            new_gen.append(new_feature2)
    fitness_sum = fitness_cache(new_gen)
    with open('fitness_dict.pkl', 'wb') as fp:
      pickle.dump(fitness_dict, fp)
    
    sorted(new_gen, key= lambda ind: fitness_dict[ind],reverse=True)
    gen = new_gen.copy()
    gen = gen[:101]
    gen_fitness = 0
    for i in range(gen):
        gen_fitness+=fitness_function(gen[i])
    file = open('output.txt', 'w+')
    file.write(str([fitness_sum,gen]))
    file.close()
    iterations+=1
    print(iterations,'-----iterations------')
  return gen


#create_individuals()

```
